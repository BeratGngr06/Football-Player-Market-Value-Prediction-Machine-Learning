```python
df["Piyasa Değeri"]=df["Piyasa Değeri"].astype(str).str.replace(".","",regex=False)
df["Piyasa Değeri"]=df["Piyasa Değeri"].astype(int)
#Veri setimizdeki piyasa değeri sütunundaki noktaları kaldırdık ve integer veri tipine dönüşmesini zorladık
```

```python
df["Yaş"]=df["Yaş"].astype(str).str.replace("(","",regex=False).str.replace(")","",regex=False)#Sayının sağındaki ve solundaki parantezleri temizledik
df["Yaş"]=pd.to_numeric(df["Yaş"],errors="coerce")#Metinleri sayısal değere çevirme errors=coerce parametre sayesinde yanlış yazılmış harf anlamsız bir şey varsa ona NaN yazıyor
yas_ort=df["Yaş"].mean()
df["Yaş"]=df["Yaş"].fillna(yas_ort)
df["Yaş"]=df["Yaş"].astype(int)




for sutun in ["Oynadığı Maç Sayısı","Attığı Gol Sayısı","Yaptığı Asist Sayısı"]:

  df[sutun]=pd.to_numeric(df[sutun],errors="coerce")
  df[sutun]=df[sutun].fillna(0)
  df[sutun]=df[sutun].astype(int) #nihai sütunu int'e çeviriyrouz.
   
    
       

```

```python
ondalikli_sutunlar=["Maç Başına Şut Sayısı","Maç Başına İsabetli Şut Sayısı","Maç Başına Dribbling","Maç Başına Top Kazanma","Maç Başına Hava Topu Kazanma","Maç Başına Başarılı Pas Sayısı","Maç Başına İsabetli Orta Sayısı"]
for sutun in ondalikli_sutunlar:
    df[sutun]=df[sutun].astype(str).str.replace(",",".",regex=False)#Virgül yerine ondalıklı sayı için nokta ile değiştirdik.
    df[sutun]=pd.to_numeric(df[sutun],errors="coerce") #hatalı verileri NaN yaptık dönüşebilenleri numeric yaptık.
    df[sutun]=df[sutun].fillna(0.0)#boş değerleri 0.0 ile doldurduk.
    df[sutun]=df[sutun].astype(float)#en sonda ise float veritipine dönüşüm

yüzdelik_sutunlar=["Pas Başarı Yüzdesi","Maç Başına İkili Mücadele Kazanma Oranı"]  
for sutun in yüzdelik_sutunlar:
    df[sutun]=df[sutun].astype(str).str.replace("%","",regex=False)# % işaretini kaldırdık.
    df[sutun]=pd.to_numeric(df[sutun],errors="coerce")
    df[sutun]=df[sutun].fillna(0.0)#Boş değerleri 0.0 ile doldurduk.
    df[sutun]=df[sutun].astype(float)

```

```python
from category_encoders.leave_one_out import LeaveOneOutEncoder

#Milliyet ve Takım sütunlarına Target Encoding uyguladık.
df_encode=df_duzenleme.copy()
encoder=LeaveOneOutEncoder(cols=["Milliyeti","Oynadığı Takım"],sigma=0.05) 
#cross valdation yapar Sigma parametresi ise oyuncuların takım ortalamasını hesapladıktan sonra veri üzerine yüzde 5 rastgele sayılar ekler çıkarır(gürültü) aynı takımda olan oyuncuların şifreli sayıları birbirinden çok az da olsa farklı olur overfitting engellenmiş olur bu sayede                                                                                        #veriyi 5'e böler Her futbolcunun takımının ortalamasını hesaplarken, o futbolcunun kendi piyasa değerini hesaplamaya dahil etmez diğer 4 parçadaki oyuncuların ortalamasına bakar. Böylece modelin sayıları birebir ezberlemesini engeller Oyuncunun kendi piyasa değerini kendi takımının ortalamasına dahil etmeyerek modelin kopya çekmesini ve ezberlemesini (leakage/overfitting) engelliyorr
                                                                                    # Smooth ise-> Az oyuncusu olan takımlarıülkelerin ortalamasını genel dünya ortalamasına doğru esnetir uç değerlerin  modeli yanıltmasının önüne geçer.
df_encode=encoder.fit_transform(df_encode,df_encode["Piyasa Değeri"])#Hem öğreniyor hem değiştiriyor.

df_encode=df_encode.rename(columns={ #Sütun isimlerini değiştirdik encoding yaptığımız anlaşılsın diye
    "Milliyeti":"Milliyeti_Encoded",
    "Oynadığı Takım":"Oynadığı Takım_Encoded"
})




                      
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
X=df_final_encoded.drop(columns=["Piyasa Değeri"])#datasetteki x ve y(hedef değişken) değişkenlerini ayrıdık
y=df_final_encoded["Piyasa Değeri"]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=36)#datasetimizi train ve test olarak ayırdık %20 test olarak

y_train_log=np.log1p(y_train) #Logaritmik dönüşüm
y_test_log=np.log1p(y_test)


print(f" Eğitim Seti oyuncu sayısı (X_train): {X_train.shape[0]} oyuncu")
print(f"Test Seti oyuncu sayısı (X_test): {X_test.shape[0]} oyuncu")


```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train) #Bağımsız değişkenleri standardizasyon yaptık (ortalaması 0 standart sapması 1)
X_test_scaled = scaler.transform(X_test)


lineer_model = LinearRegression() #Örneklem aldık
lineer_model.fit(X_train_scaled, y_train_log) #Modeli eğittik burda 

lineer_model_prediction_log = lineer_model.predict(X_test_scaled) #burda teste tabi tuttuk modelimizi
#Logaritmik dönüşüm yaptığımız için tahminler de logaritmik ölçüde

lineer_model_r2_skor = r2_score(y_test_log, lineer_model_prediction_log)#eğittiğimiz modeli r2 skoruna bakıyoruz(gerçek veriler ve tahmin verileri ile)

prediction_gercek = np.expm1(lineer_model_prediction_log)#burada logaritmik dönüşümü geriye aldık.
y_test_gercek = np.expm1(y_test_log) 

lineer_model_mae = mean_absolute_error(y_test_gercek, prediction_gercek) #ortalama mutlak hata metriği hesaplama 
lineer_model_rmse = np.sqrt(mean_squared_error(y_test_gercek, prediction_gercek))

lineer_model_n=X_test.shape[0]#test versinideki oyuncu sayısı
lineer_model_p=X_test.shape[1]#test verisindeki sütun sayısı
lineer_model_adjusted_r2=1-(1-lineer_model_r2_skor*(lineer_model_n-1)/(lineer_model_n-lineer_model_p-1))#adjusted r2 hesabı

lineer_model_r2_skor_gercek=r2_score(y_test_gercek,prediction_gercek)

lineer_model_cross_val_score=cross_val_score(lineer_model,X_train_scaled,y_train_log,cv=5,scoring="r2")#cross validation hesabı modelin şans faktörüne bağlı olup olmadığını anlamak için

print(" --BAŞARI METRİKLERİ--")
print(f"R2 Skor: {round(lineer_model_r2_skor, 4)}")
print(f"MAE (Mean Absolute Error): {round(lineer_model_mae, 2)} Euro")#8.5 milyon € civarında bir sapma var
print(f"RMSE (Root Mean Squared Error): {round(lineer_model_rmse, 2)} Euro") #yaklaşık 21.4 milyon € 
print(f"Adjusted R2 skor:{round(lineer_model_adjusted_r2,4)}")#işe yaramayan sütun eklendiğinde adjusted r2 düşebilir ama r2 düşmez
print(f"R2 skor Logaritmik dönüşümsüz:{round(lineer_model_r2_skor_gercek,4)}")

print("--Cross Validation Analizi--")
print(f"5 farklı katlama R2 Skorları:{np.round(lineer_model_cross_val_score,4)}")#cv sonucu r2 skorlarını ekrana bastırma
print(f"Ortalama Cross validation R2 skoru:{round(lineer_model_cross_val_score.mean(),4)}")#cv sonucu ortalama r2 skoru
print(f"Skorların standart sapması:{round(lineer_model_cross_val_score.std(),4)}")#cv sonucu r2 skorların standart sapması
```
